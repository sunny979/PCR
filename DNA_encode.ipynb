```python
# 将字节数组转换成01序列
def DecToBin(num:int):
    result = []
    while(num // 2 != 0):
        result.append(str(num % 2))
        if(num // 2 !=0):
            num //= 2
    if num == 1:
        result.append("1")   
    while(len(result) < 8):
        result.append("0")
    result.reverse()  
    return "".join(result)


def convert_byte_array_to_01(byte_array):
    result = ""    
    for byte in byte_array:
        result += DecToBin(byte)    
    return result

def read_to_01(name):
    f = open(name, 'rb')
    t = f.read()
    t=convert_byte_array_to_01(t)    
    return t
```

```python
def Total(string,count):
    dt={}
    Four=[]                  #生成4的几次方各种数值，目前是到4的19次方
    for i in range(20):
        Four.append(pow(4,i))
        
    for i in range(3,count):
        k,L,items=GAP_total(string,i)
        dt[k]=L
   
    Storage_Density={}
    for item in dt.items():
        key=int(item[0])
        value=int(item[1])
        for ch in Four:
            if value< int(ch):
                DNA_bit=Four.index(ch)
                single_density=round(int(item[0])/DNA_bit,2)
                n0=len(string)
                if n0%key != 0:
                    nstr=string
                    nstr += "0" * (key-n0%key)
                    n1=len(nstr)
                else:
                    nstr=string
                    n1=len(nstr)
                encode_len=int(n1/key*DNA_bit)
                codebook_len=(key+DNA_bit)*value
                all_len=encode_len+codebook_len
                all_density=round(n0/all_len,2)
                '''
                print('encode_len',encode_len)
                print('codebook_len',codebook_len)
                print('all_len',all_len)
                '''
                encode=( DNA_bit,single_density,encode_len,codebook_len,all_density)
                Storage_Density[item]=encode
                break


    lx=[]
    ly=[]
    for item in Storage_Density.items():
        lx.append(item[0][0])
        ly.append(item[1][4])  
    
    plt.plot(lx, ly)
    plt.show()
    
    storage=list(Storage_Density.items())
    storage.sort(key=lambda x:x[1][4],reverse=True)
    #print(storage)
    
    
    return storage
```

```python
def filter_primer_list(raw_set: set, GAP:int)->list:
    #print("原始集合大小：" , len(raw_set))
    bases = ["A", "T", "C", "G"]
    min_range = 5
    assert min_range <= GAP
    raw_list = list(raw_set)
    for idx in range(len(raw_list)-1, -1, -1):
        p = raw_list[idx]
        for base in bases:
            if p not in raw_list:
                break
            
            if base*min_range in p:
                raw_list.remove(p)

            
    #print("删除后集合大小：" , len(raw_list))
    return raw_list
```

```python
def total_primer(GAP):
    global pri
    primer_dict = {}
    
  
    for i in range(len(pri)):
        for j in range(len(pri[i][-8:])-GAP+1):
            t = pri[i][j:j+GAP]
            primer_dict[t]=primer_dict.get(t,0)+1
            
    primer_ls=list(primer_dict.items())
    primer_ls.sort(key=lambda x:x[1],reverse=True)

    primer_dict_set = set()
    for p in primer_ls:
        primer_dict_set.add(p[0])
        
    #print(len(primer_dict_set))
    
    seq = SequenceIterator.SeqGenerator(GAP, 1)
    seq = iter(seq)
    total_set = set()
    for i in range(4**GAP):
        total_set.add(next(seq))
    remain_set = total_set-primer_dict_set
    
    if len(remain_set) != 0:
        filter_list = filter_primer_list(remain_set, GAP)
    
        for p in filter_list:
            primer_dict[p] = 0
        
    primer_ls=list(primer_dict.items())
    primer_ls.sort(key=lambda x:x[1],reverse=True)
    
    
    return primer_ls
```

```python
def A_00_encode(seq):
    n0=len(seq)
    if n0%2 != 0:
        seq1=seq
        seq1+='0'
    else:
        seq1=seq
    n1=len(seq1)
    seq_to_ATCG=''
    for i in range(0,n1,2):
        if seq1[i:i+2]=='00':
            seq_to_ATCG+='A'
        elif seq1[i:i+2]=='01':
            seq_to_ATCG+='T'
        elif seq1[i:i+2]=='10':
            seq_to_ATCG+='C'
        else:
            seq_to_ATCG+='G'
            
    return seq_to_ATCG
```
