```python
from copy import deepcopy
import time
```

```python
f=open('FinalPrimer.txt','r').read()
f = eval(f)
primer_8 = []
for primer in f:
    primer_8.append(primer[-8:])
```

```python
def check_need_jiaozhi(seq:str):
    for pri in primer_8:
        if pri in seq:
            # print(seq.index(pri))
            return False
        
    return True
```

```python
def jz(seq):
    slist=[]
    for i in range(0,225,15):
        slist.append(seq[i:i+15])   # 将序列变成列表
    
        
    jz = [[""]*15 for i in range(15)]  #生成15*15的二维空列表
    
    for i in range(15):
        for j in range(15):
            jz[i][j]=slist[i][j]   #填充交织器
    
    s1=''                             #交织1
    for j in range(9):  
        for i in range(7):     
            s1+=jz[6-i][j]
            
    s2=''                             #交织2
    for j in range(1,7):
        for i in range(8):     
            s2+=jz[7-i][-j]

    s3=''                            #交织3
    for j in range(0,9,2):
        for i in range(8):     
            s3+=jz[-i-1][j]
    for j in range(1,9,2):
        for i in range(8):     
            s3+=jz[-i-1][j]
    
    s4=''                            #交织4
    for j in range(14,8,-2):
        for i in range(7):     
            s4+=jz[-i-1][j]
    for j in range(13,8,-2):
        for i in range(7):     
            s4+=jz[-i-1][j]
    
    s_jz=s1+s2+s3+s4          #4个交织器结果连在一起
    
    return s_jz
```

```python
def generate_good_seq(seq):
    time_start = time.time()  # 记录开始时间
    global primer_8
    count = 0
    satisfy = False
    max_loop = 500
    while not satisfy and count < max_loop:
        
        satisfy = check_need_jiaozhi(seq)
        if satisfy:
            time_end = time.time()  # 记录结束时间
            time_sum = time_end - time_start  # 计算的时间差为程序的执行时间，单位为秒/s
            time_sum = round(time_sum,6)
            print('交织过程所用时间：',time_sum,'秒')
            return seq, count
        else:
            seq = jz(seq)
            # print(seq)
            count += 1
            continue 
    
    return "一直都重复"
```
